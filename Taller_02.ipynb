Construya un DataFrame a partir del siguiente diccionario

`exam_data = {'name': ['Anastasia', 'Dima', 'Katherine', 'James', 'Emily', 'Michael', 'Matthew', 'Laura', 'Kevin', 'Jonas'], 'score': [12.5, 9, 16.5, 10, 9, 20, 14.5, 12, 8, 19], 'attempts': [1, 3, 2, 3, 2, 3, 1, 1, 2, 1], 'qualify': ['yes', 'no', 'yes', 'no', 'no', 'yes', 'yes', 'no', 'no', 'yes']}`

con las siguientes etiquetas para los índices

`labels = ['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i', 'j']`

1. Seleccione solo las columnas *name* y *score*.
  
2. Filtre las filas donde el número de intentos (attemps) sea mayor que 2.

3. Filtre las filas donde el *score* se encuentre entre 15 y 20 (incluidos)

4. Cambie el *score* en la fila 'd' a 11.5

5. Agregue una nueva fila con etiqueta 'k' al dataframe con un valor que usted desee para cada columna

6. Borre la fila 'k' y entregue nuevamente el DataFrame original.

7. Ordene el DataFrame por nombre en orden ascendente.

8. Reemplace los valores de la columna *qualify* que contiene los valores *yes* y *no* por los valores booleanos True y False respectivamente.

### **Solucion**

```python
#Importación de librerias a utilizar
import pandas as pd
import numpy as np
```

##### Construcción del Dataframe

```python
# Diccionario con los datos
exam_data = {'name': ['Anastasia', 'Dima', 'Katherine', 'James', 'Emily', 'Michael', 'Matthew', 'Laura', 'Kevin', 'Jonas'],
             'score': [12.5, 9, 16.5, 10, 9, 20, 14.5, 12, 8, 19],
             'attempts': [1, 3, 2, 3, 2, 3, 1, 1, 2, 1],
             'qualify': ['yes', 'no', 'yes', 'no', 'no', 'yes', 'yes', 'no', 'no', 'yes']}

# Etiquetas para los índices
labels = ['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i', 'j']


# **1. Documentación:** 

"""
DataFrame.

Columnas:
    - name: Nombre del estudiante.
    - score: Puntuación del examen.
    - attempts: Número de intentos realizados.
    - qualify: Indica si el estudiante aprobó ('yes') o no ('no').

Índice:
    - Etiquetas personalizadas ('a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i', 'j').
"""
```

```python
# Creación del DataFrame
df = pd.DataFrame(exam_data, index=labels)
```

```python
display(df)
```

1. Seleccione solo las columnas *name* y *score*.

```python
# Seleccionar las columnas "name" y "score"
df_selected = df[['name', 'score']]

# Visualizar las primeras filas del DataFrame seleccionado
display(df_selected.head())
```

2. Filtre las filas donde el número de intentos (attemps) sea mayor que 2.

```python
# Filtrar filas donde el número de intentos es mayor a 2
df_filtered = df[df['attempts'] > 2]

# Visualizar las filas del DataFrame filtrado
display(df_filtered)
```

3. Filtre las filas donde el *score* se encuentre entre 15 y 20 (incluidos)

```python
# Filtrar filas donde el "score" está entre 15 y 20 (inclusive)
df_filtered = df[df['score'].between(15, 20)]

# Visualizar las filas del DataFrame filtrado
display(df_filtered)
```

4. Cambie el *score* en la fila 'd' a 11.5

```python
# Cambiar el valor de "score" en la fila 'd' a 11.5
df.loc['d', 'score'] = 11.5

# Visualizar la fila 'd' del DataFrame
print(df.loc['d'])
```

5. Agregue una nueva fila con etiqueta 'k' al dataframe con un valor que usted desee para cada columna

```python
# Agregar una nueva fila con etiqueta 'k'
nueva_fila = {'name': 'Ana', 'score': 18.0, 'attempts': 2, 'qualify': 'yes'}
df.loc['k'] = nueva_fila  
```

```python
display(df)
```

6. Borre la fila 'k' y entregue nuevamente el DataFrame original.

```python
# Eliminar la fila 'k'
df = df.drop('k')

# Visualizar el DataFrame
display(df)
```

7. Ordene el DataFrame por nombre en orden ascendente.

```python
# Ordenar el DataFrame por "name" en orden ascendente
df = df.sort_values('name')

# Visualizar el DataFrame
display(df)
```

8. Reemplace los valores de la columna *qualify* que contiene los valores *yes* y *no* por los valores booleanos True y False respectivamente.

```python
# Reemplazar "yes" por True y "no" por False en la columna "qualify"
df['qualify'] = df['qualify'].replace({'yes': True, 'no': False})

# Visualizar el DataFrame
display(df)
```

## Ejercicio 2 (30%)
Importe el dataset `athletes` del repositorio del curso y realice las siguientes tareas:

1. Importarlo como nombre athletes
2. Filtre el dataframe tal que solo se quede con los atletas que hayan ganado al menos una medalla.
4. Calcule la masa muscular por fila con una función como

$$ I_{m.m.}=\frac{Weight}{Height}^2 $$

4. Agrupe los datos por calculando la media por deporte para cada una del resto de características. ¿Cuál es el deporte en el que los jugadores tienen mayor peso promedio?

5. ¿Cuál es el pais con más participación femenina?

### **Solucion**

```python
# 1. Importarlo como nombre athletes
athletes = pd.read_csv("https://raw.githubusercontent.com/tomasate/Datos_Clases/main/Datos_1/athletes.csv")
athletes
```

```python
#2.  Filtre el dataframe tal que solo se quede con los atletas que hayan ganado al menos una medalla.

# Seleccionar las columnas con los tipos de medallas para sumarlas
column_names = ['gold','silver','bronze']

# Se suman las medellas y se crea la variable 'total_medal'
# que corresponde al total de medallas ganadas por cada deportista
athletes['total_medal'] = athletes[column_names].sum(axis=1)

#se hace un filtro de los deportistas que hayan ganado al menos una medalla.
# por lo que el filtro queda para total_medal mayor que 0.
# se organiza la tabla para observar que atleta fue el que más medallas ganó.
athletes_medals = athletes.loc[athletes['total_medal']>0]\
                          .sort_values(by=['total_medal'],ascending=False)
```

```python
# 3. Calcule la masa muscular por fila

# se calcula el indice de mása muscular IMC
athletes['imc']= athletes['weight']/(pow(athletes['height'],2))

#se muestra un breve resumen de como se distribuye la variable.
athletes['imc'].describe()
```

```python
# 4. Agrupe los datos por calculando la media por deporte para cada una del resto de características.
#    ¿Cuál es el deporte en el que los jugadores tienen mayor peso promedio?

# Se elimina la columna id para eviitar resumirla, ya que no aporta valor.
# Se agrupa por deporte.
# Se calcula el promedio por todas las variables númericas agrupada por deporte
# Se organiza de manera descendete para extraer el top 1 del deporte con mayor peso promedio.
summary_athletesBySport = athletes.drop('id',axis = 1)\
                                  .groupby('sport')\
                                  .mean()\
                                  .sort_values(by='weight',ascending=False)

# Se imprime el resultado del deporte con mayor peso promedio.
print(summary_athletesBySport.head(1))

#Respuesta:  El deporte donde los jugadores tienen el mayor peso promedio es Basketball
```

```python
# 5. ¿Cuál es el pais con más participación femenina?

# Se hace un filtro en el que solo se incluyen a las muejeres.
# Se agrupa por nacionalidad.
# se determina la cantidad de mujeres por nacionalidad.
# Se organiza descendentemente para extaer el top 1 de país con mayor mujeres participantes.
summary_1 = athletes_medals[athletes_medals['sex']== 'female']\
                           .groupby(['nationality'])\
                           .size()\
                           .sort_values(ascending=False)

# se imprime el resultado del país top 1 con más mujeres participantes
print(summary_1.head(1))
# Respuesta: el país con mayor participación es USA
# con 117 mujeres participando y ganando mendallas.
```

## **Ejercicio 3 (30%)**

Importe el dataset de nba del repositorio de clase y responda (hint: Elimine las instancias con registros faltantes):

1. Cuáles son los equipos que pagan los salarios más altos (presente el resultado como un dataFrame con el nombre del equipo y la cantidad de dinero de la nómina)
2. Separe la variable *Age* en tres grupos y cree una nueva columna llamada *Age-Group* con esta información.<
3. Construya una tabla dinámica que muestre el salario promedio de los jugadores por equipo y posición para para los cinco equipos que pagan mejores salarios.

### **Solucion**

```python
# importamos la base de datos del repositorio GIT
nba_real = pd.read_csv("https://raw.githubusercontent.com/tomasate/Datos_Clases/main/Datos_1/nba.csv")
nba_real
```

```python
# debido a que la vamos a modificar hacemos una copia de la misma
nba = nba_real.copy()
# se pasa el nombre de las columnas a minuscula, para facilitar el trabajo más adelante
nba.columns = nba.columns.str.lower()

nba
```

```python
# borro la columna "unnamed: 0" y "collage" la primera para que no confunda con el indice y la segunda debido a que tiene datos faltantes, sin embargo dichos datos faltantes no son impedimento para el analisis
# que se solicita en este ejercicio, por tanto se decide borrar la columna antes de borrar las filas con datos faltantes.
nba = nba.drop(['unnamed: 0','college'], axis = 1)
#borrado de filas con datos faltantes
nba = nba.dropna(axis = 0)
nba
```

```python
# Se realiza una agrupación por equipo, manteniendo un indice distinto a "equipo", obteniendo los valores totales de el pago de salarios por equipo (sum())
df = nba.groupby('team', as_index = False)[['salary']].sum()
df
```

```python
# se organiza la tabla en orden decendiente y se obtienen los primeros 5 resultados, así mismo se resetea el indice para que el mismo quede acorde al nuevo orden de la tabla, por ultimo
# se dejan unicamente los primeros 5 valores correspondientes a los 5 equipos que más pagan nómina.
top_teams = df.sort_values(by = 'salary', ascending = False).reset_index(drop = True)[:5]

top_teams
```

```python
# lo primero que se debe hacer para solucionar el numeral B es tener la información de la edad de todos los jugadores.
nba.age.describe()
```

```python
#con está información podemos conculuir que la diferencia de edad entre el jugador mayor y el menor es de 21 años por lo que se dividirá en 3 grupos de 7 años cada uno así
# grupo a: (> 18 <= 26) años, grupo b: (> 26 <= 33) grupo c: (> 33 <= 40)

nba['Agegroup'] = pd.cut(nba.age, [18.0, 26.0, 33.0, 40.0], labels = ["a", "b", "c"])
nba
```

```python
# así pues, el anterior arreglo da cuenta de la solución del punto B. como se muestra en la siguiente descripción cuenta todas las filas con las que se trabajo en el numeral A.
nba.Agegroup.describe()
```

```python
#para el numeral C se toma el resultado obtenido en A llamado top_teams, "filtrando" los valores de nba que esten en la columna 'team' de  "top_teams"

salario_promedio = nba[nba.team.isin(top_teams['team'])]

# se grafica la tabla de salarios (promedio) de los primeros 5 equipos teniendo como indice el equipo y como columnas la posición de cada jugdor.
# aggfunc = 'mean' es redundante dado que automaticamente se hace con esa función, se deja para facilitar algún cambio.

salario_promedio.pivot_table('salary', index = 'team', columns= 'position', aggfunc='mean')
```

```python
# ahora a manera de prueba se buscará de forma "manual" los resultados para "Chicago Bulls" y posición "C" al tiempo
# sacando el promedio de dicho filtrado se obtiene el mismo resultado que se obtiene en la tabla
nba.loc[(nba['team'] == 'Chicago Bulls') & (nba['position'] == 'C'), 'salary'].mean()
```
